```python
import pandas as pd

from dostoevsky.models import FastTextSocialNetworkModel
from dostoevsky.tokenization import RegexTokenizer

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pymorphy2

nltk.download('stopwords')
nltk.download('punkt')
```

```python
songs_df = pd.read_pickle('data_my/songs_tagged.pkl')
songs_df.head()
```

```python
tokenizer = RegexTokenizer()
model = FastTextSocialNetworkModel(tokenizer=tokenizer)
sentecnes = songs_df['lyrics'].to_list()
predicted = model.predict(sentecnes)
```

```python
predicted
```

```python
res = []
for i in range(len(songs_df)):
    row = songs_df.iloc[i]
    prediction = predicted[i]
    d = {
        'title': row['title'],
        'album_id': row['album_id'],
        'positive': prediction['positive'],
        'negative': prediction['negative'],
        'neutral': prediction['neutral']
    }
    res.append(d)
    
df_sentiment = pd.DataFrame(res)
df_sentiment
```

```python
df_sentiment.to_csv('data_my/kish_sentiment.csv',index=False, encoding='UTF-8')
```

```python
morph = pymorphy2.MorphAnalyzer()

russian_stopwords = set(stopwords.words('russian'))
additional_stopwords = {'и', 'припев', 'но', 'я', 'в', 'но', 'что', 'мой', 'свой', 'весь', 'всё', 'на', 'мы', 'c', 'a','вест','это', 'сам'}
russian_stopwords.update(additional_stopwords)
tokenizer = RegexTokenizer()



def tokenize_and_base(text):
    words = tokenizer.split(text)
    processed_words = []
    for word in words:
        # Remove punctuation
        if not word.isalpha():
            continue
        if word in russian_stopwords:
            continue
        # lemmatiza        
        parsed_word = morph.parse(word)[0]
        normal_form = parsed_word.normal_form
        if normal_form in russian_stopwords:
            continue
        processed_words.append(normal_form)
```
